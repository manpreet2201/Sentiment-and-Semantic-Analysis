```python
users_locs = [{"full_text" : tweet.full_text } for tweet in tweets_data]
```

```python
import re
cleaned_data=[]
#reference https://programmersought.com/article/15261349366/
def clean_emoji(string):
    pattern = re.compile("\["
                         u"\U0001F600-\U0001F64F"  # emoticons
                         u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                         u"\U0001F680-\U0001F6FF"  # transport & map symbols
                         u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                         u"\U00002702-\U000027B0"
                         u"\U000024C2-\U0001F251"
                         "\]+", flags=re.UNICODE)
    return pattern.sub(r'', string)

# removing special characters & URLs
# special characters regex reference: https://www.webdeveloper.com/d/199621-need-a-regex-to-exclude-all-but-a-za-z0-9s
# URl cleaning regex
for tweet_detail in users_locs:
    tweet_detail['full_text'] = clean_emoji(tweet_detail['full_text'])
    tweet_detail['full_text'] = re.sub(r'[^a-zA-Z0-9\s\.]+', '', tweet_detail['full_text'])
    tweet_detail['full_text'] = re.sub(r'http\S+', '', tweet_detail['full_text'], flags=re.MULTILINE)
    cleaned_data.append(tweet_detail)

cleaned_data
```

```python
#reference https://stackabuse.com/python-for-nlp-creating-bag-of-words-model-from-scratch/
import nltk
nltk.download('punkt')
bagofwords=[]

for sentence in cleaned_data:
    wordfreq = {}
    tokens = nltk.word_tokenize(sentence.get("full_text"))
    for token in tokens:
        if token not in wordfreq.keys():
            wordfreq[token] = 1
        else:
            wordfreq[token] += 1
    bagofwords.append(wordfreq);

    
```
